### 6. 特徴量のうちカテゴリ変数について、ユニークな値を取得する

### 8. 特徴量のうち量的変数について、目的変数との関係を可視化する

```
dfCorr = pd.DataFrame([], columns = ["corr"])
for column in dfTips.columns :
    if column != "tip" :
        if str(dfTips.loc[:,column].dtypes) not in ["object", "category"]  :
            corr = np.corrcoef(dfTips[column], dfTips["tip"])[0, 1]
            ser = pd.Series([corr], index=dfCorr.columns, name=column )
            dfCorr = dfCorr.append( ser )
print( dfCorr )
```

```
dfCorr = pd.DataFrame([], columns = ["corr"])
for column in dfBoston.columns :
    if column != "MEDV" :
        if str(dfBoston.loc[:,column].dtypes) not in ["object", "category"]  :
            corr = np.corrcoef(dfBoston[column], dfBoston["MEDV"])[0, 1]
            ser = pd.Series([corr], index=dfCorr.columns, name=column )
            dfCorr = dfCorr.append( ser )
print( dfCorr )
```

### 9. 量的変数について、データの分布を可視化する

```
numericCol = []
for column in dfTips.columns :
    if str(dfTips.loc[:,column].dtypes) not in ["object", "category"]  :
        numericCol.append( column )

f = pd.melt(dfTips, value_vars=numericCol)
g = sns.FacetGrid(f, col="variable",  col_wrap=3, sharex=False, sharey=False, size=5)
g = g.map(sns.distplot, "value")
```

```
numericCol = []
for column in dfTitanic.columns :
    if str(dfTitanic.loc[:,column].dtypes) not in ["object", "category"]  :
        if column not in ["PassengerId", "Pclass"] :
            numericCol.append( column )

f = pd.melt(dfTitanic, value_vars=numericCol)
g = sns.FacetGrid(f, col="variable",  col_wrap=3, sharex=False, sharey=False, size=5)
g = g.map(sns.distplot, "value")
```

```
numericCol = []
for column in dfBoston.columns :
    if str(dfBoston.loc[:,column].dtypes) not in ["object", "category"]  :
        numericCol.append( column )

f = pd.melt(dfBoston, value_vars=numericCol)
g = sns.FacetGrid(f, col="variable",  col_wrap=3, sharex=False, sharey=False, size=5)
g = g.map(sns.distplot, "value")
```
